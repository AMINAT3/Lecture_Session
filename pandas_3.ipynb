```python
def  male_or_female(value_from_series):
    
    """
    this is a doctring...
    parameter: a value from the sex_upon_outcome series
    in the Austine Animal Shelter dataset'
    There are five possible values:
    'Spayed Femal', 'Unknown', 'Intact Female', 'Intact Male', 
        'Neutered Male'
        
        Returns:
        female, male, unknoown
        
        """
    if 'female' in value_from_series.lower():
        return 'female'
    
    elif ' male' in value_from_series.lower():
        return 'male'
    
    else:
            return 'unknown'
        
        
animal_df['sex_upon_outcome'].apply(male_or_female)
```

```python
# we can now make a new column, sex:

animal_df['sex'] = animal_df['sex_upon_outcome'].apply(male_or_female)
animal_df.head()
```

```python

def age_to_days(age):
    
    """
    params: age upon outcome of shelter animal.
    A number followed by a unit of time
    'NULL', 'days', 'months', 'weeks', 'yar', 'years'
    
    returns days old outcome
    """
    
    age_split = age.split(' ')
    
    if len(age_split) == 1:
        return 'NULL'
    
    elif age_split[1] == 'days':
        return age_split[0]
    
    elif age_split[1] == 'month':
        return str(int(age_split[0])*30)
    
    elif age_split[1] == 'year':
        return str(int(age_split[0])*365) + 'days'
    
    else:
        pass
    
animals['days_upon_outcome'] = animals['age_upon_outcome'].map(age_to_days)

animals['days_upon_outcome']
```
